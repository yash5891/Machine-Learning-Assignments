**Grouped Bar Chart**

```python
CourseTaught = ["M.Sc CS", "M.Sc. DS", "M.SC CA", "IMCA"]
Girls = [25, 12, 24,11]
Boys = [40,22, 40, 20]
x = np.arange(len(CourseTaught))
width = 0.20
#plot x, y values
fig, ax = plt.subplots()
bar1 = ax.bar(x - width/2, Boys, width, label = "Boys")
bar2 = ax.bar(x + width/2, Girls, width,
label = "Girls")
ax.set_xticks(x)
ax.set_xticklabels(CourseTaught)
ax.legend()
plt.show()
```

1)Set the figure size and adjust the padding between and around the subplots.

2)Make a dataframe, df, of two-dimensional, size-mutable, potentially heterogeneous tabular data.

3)Make an ordered dataframe, using sort_values().

4)Make a list in the range of dataframe index.

5)Create a stem plot, using the ordered dataframe.

6)Set xticks and labels using xticks() method.

7)To display the figure, use show() method.





```python
import pandas as pd
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True
df = pd.DataFrame({'group': list(map(chr, range(65, 85))),
                   'values': np.random.uniform(size=20)})

ordered_df = df.sort_values(by='values')
my_range = range(1, len(df.index)+1)
plt.stem(ordered_df['values'])
plt.xticks(my_range, ordered_df['group'])
plt.show()
```

**Area Charts**

```python
data = {'Year': [2015, 2016, 2017, 2018, 2019],
        'Sales': [100, 150, 200, 180, 250]}
df = pd.DataFrame(data)
df.head()
```

```python
#stacked Area Charts
import pandas as pd

import plotly.express as px

# Create a dataframe with random values

data = pd.DataFrame({

    'Year': [2015, 2016, 2017, 2018, 2019],

    'Category A': [10, 20, 30, 40, 50],

    'Category B': [20, 30, 40, 50, 60],

    'Category C': [30, 40, 50, 60, 70]

})

# Create a stacked area chart

fig = px.area(data, x='Year', y=['Category A', 'Category B', 'Category C'], title='Stacked Area Chart')

fig.show()
```

**Waffle Charts**

```python
import pandas as pd
import matplotlib.pyplot as plt
from pywaffle import Waffle

# creation of a dataframe
data ={'phone': ['Xiaomi', 'Samsung',
                 'Apple', 'Nokia', 'Realme'],
       'stock': [44, 12, 8, 5, 3]
     }

df = pd.DataFrame(data)

# To plot the waffle Chart
fig = plt.figure(
    FigureClass = Waffle,
    rows = 5,
    values = df.stock,
    labels = list(df.phone)
)
```
