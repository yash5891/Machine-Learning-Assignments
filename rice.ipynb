```python
# Assuming you have a dataframe named 'Rice' containing the rice data

# Get a list of unique states
states = Rice['state'].unique()

# Iterate over each state
for state in states:
    # Create a dataframe for the current state
    state_df = Rice[Rice['state'] == state]

    # Display the first few rows of the dataframe
    print(f"\n{state} DataFrame:")
    print(state_df.head())

    # Calculate the length of the 'market' column
    print(f"Number of markets in {state}: {len(state_df['market'])}")

    # Calculate and display the maximum and minimum values of 'min_price' and 'max_price'
    print(f"Maximum min_price in {state}: {max(state_df['min_price'])}")
    print(f"Minimum min_price in {state}: {min(state_df['min_price'])}")
    print(f"Minimum max_price in {state}: {min(state_df['max_price'])}")
    print(f"Maximum max_price in {state}: {max(state_df['max_price'])}")
```
