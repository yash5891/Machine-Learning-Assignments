```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#last 5 rows
df=pd.read_csv(r"/content/onlinefoods.csv")
df.tail()
df.head()
```

```python
# @title Marital Status vs Monthly Income

df.groupby(['Monthly Income', 'Marital Status']).size().unstack().plot(kind='bar', stacked=True)
```

```python
#columns detail
df.info()
```

```python
#statistical information
df.describe()
```

```python
#4#orders of food married or single
order_status=df["Marital Status"].value_counts()

status_counts = order_status


plt.figure(figsize=(8, 6))
plt.pie(status_counts,  autopct='%1.1f%%', startangle=45)


plt.title('Distribution of food orders based on family size')
plt.axis("on")


plt.show()
```
